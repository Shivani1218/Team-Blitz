```python
import pandas as pd
```

```python
df=pd.read_csv('stp.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder#preprocessing
enc = LabelEncoder()
enc.fit(df.gender)
df.gender = enc.transform(df.gender)
enc.fit(df.race)
df.race = enc.transform(df.race)
enc.fit(df.parental)
df.parental = enc.transform(df.parental)
enc.fit(df.lunch)
df.lunch = enc.transform(df.lunch)
enc.fit(df.test_preparation_course)
df.test_preparation_course = enc.transform(df.test_preparation_course)
df
```

```python
x=df.iloc[:,[0,1,2,3,4]]
y=df.iloc[:,[5,6,7]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtreg1=DecisionTreeRegressor()
dtreg1.fit(xtrain,ytrain)
ypred_1=dtreg1.predict(xtest)
ypred_1
```

```python
dtreg1.score(xtest,ytest)*100
```
